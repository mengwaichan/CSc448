```python
import matplotlib.pyplot as plt
import numpy as np

# Define the models and corresponding performance metrics
models = ['BIDIRECTIONAL LSTM', 'CNN+LSTM', 'CNN', 'MNB', 'Decision Tree', 'Random Forest', 'SVM', 'Logistic Regression']
accuracy = [92.882206, 90.576441, 92.932331, 82.6566, 88.07017543859649, 90.17543859649123, 87.76942355889724, 87.92]
precision = [92.882206, 90.576441, 92.932331, 83.6566, 88.1089951820483, 90.16448509720978, 88.06868101778665, 88.13]
recall = [92.882206, 90.576441, 92.932331, 82.6566, 88.07017543859649, 90.17543859649123, 87.76942355889724, 87.92]
f1 = [92.882206, 90.576441, 92.932331, 82.6566, 88.07017543859649, 90.17543859649123, 87.76942355889724, 87.59]

# Create an array of indices for the models
x = np.arange(len(models))

# Set up the plot figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.2

# Create bar plots for each performance metric
rects1 = ax.bar(x - 1.5 * bar_width, accuracy, bar_width, label='Accuracy', color='#1f77b4')
rects2 = ax.bar(x - 0.5 * bar_width, precision, bar_width, label='Precision', color='#ff7f0e')
rects3 = ax.bar(x + 0.5 * bar_width, recall, bar_width, label='Recall', color='#2ca02c')
rects4 = ax.bar(x + 1.5 * bar_width, f1, bar_width, label='F1', color='#d62728')

# Set x-axis ticks and labels
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Label the axes and title of the plot
plt.xlabel('Models')
plt.ylabel('Performance')
plt.title('Model Comparison')

# Set a more zoomed-in y-axis limit for better visualization
plt.ylim(80, 95)

# Display the plot
plt.show()


```
